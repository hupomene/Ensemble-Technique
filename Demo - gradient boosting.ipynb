```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
from sklearn.ensemble import GradientBoostingRegressor #statquest - youtube
```

### The math:

$ \text{Actual values} = y_i $

$ \text{Predictions} = y_i^p $  

$ \text{Loss function} = L(y_i, y_i^p) $  

$ \text{Loss function} = MSE = \sum {(y_i - y_i^p)}^2 $  

$ \text{Derivative of loss function} = \delta {L(y_i, y_i^p)}/ \delta{y_i^p } $ 

$ y_i^p = y_i^p + \alpha * \delta {L(y_i, y_i^p)}/ \delta{y_i^p } $  

$ y_i^p = y_i^p + \alpha * \delta {\sum {(y_i - y_i^p)}^2}/ \delta{y_i^p } $  

$ y_i^p = y_i^p - \alpha * 2*{\sum {(y_i - y_i^p)}} $  

```python
x = np.arange(0,50)
x = pd.DataFrame({'x':x})

# random uniform distributions in differnt ranges

y1 = np.random.uniform(10,15,10)
y2 = np.random.uniform(20,25,10)
y3 = np.random.uniform(0,5,10)
y4 = np.random.uniform(30,32,10)
y5 = np.random.uniform(13,17,10)

y = np.concatenate((y1,y2,y3,y4,y5))
y 
```

```python
marker_size = 12
medium_size = 28

plt.rc('axes', labelsize=medium_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=medium_size)    # fontsize of the tick labels
plt.rc('ytick', labelsize=medium_size)    # fontsize of the tick labels
plt.rc('lines', markersize=marker_size)    # fontsize of the tick labels

fig, ax = plt.subplots(figsize=(10,5))
# plt.scatter(x, y)
plt.plot(x,y, 'go', alpha=0.7)
ax.set_xlabel('x')
ax.set_ylabel('y')
```

### Number estimators = 1
<img src="image1.png">

```python
gbr_1 = GradientBoostingRegressor(n_estimators = 1, learning_rate = 1, max_depth=1, random_state=22)

gbr_1.fit(x, y)
y_predict_1 = gbr_1.predict(x)
```

```python
y_predict_1
```

```python
residuals_1 = y - y_predict_1
```

```python
residuals_1
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize = (13,5))

ax1.plot(x,y, 'go', alpha=0.7, label="y")
ax1.plot(x, y_predict_1, 'r', label="y_pred")
ax1.set_title('Prediction (1 tree)', fontsize=medium_size)
ax1.set_xlabel('x')
ax1.legend(loc='upper left')

ax2.plot(x, residuals_1, 'ro', alpha=0.7)
ax2.set_title('Residuals vs. x (1 tree)', fontsize=medium_size)
ax2.set_xlabel('x')
ax2.set_ylabel('Residuals')
```
